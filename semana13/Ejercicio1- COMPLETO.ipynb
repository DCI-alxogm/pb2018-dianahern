```python
##Aquí declaro mis librerías que voy a usar
import matplotlib.pyplot as plt ## Esta librería sirve para generar gráficos
import numpy as np ##Esta librería sirve para poner arreglos y funciones matemáticas 
%matplotlib inline 
```

Ejercicio 1: Círculos con puntos
Se mostrará el progreso paso por paso

```python
i=np.random.rand(10) ##La función np.ramdom sirve para que se escojan 10 puntos de forma aleatoria
j=np.random.rand(10)
plt.scatter(i,j) ##Esta función imprime los 10 puntos, en las coordenadas dadas por [i,j]
z=np.array([i,j]) ##Convierto cada coordenada en un arreglo con el nombre de "s"
##Hago esto para checar cómo funciona el arreglo "z"
print (z)
print (len(z)) ##Checo las filas que tiene "z"
print (z[1,1]) ##Imprimo y checo cómo está distribuida mi matriz
```

```python
##Este FOR empieza en 0 y termina en 360, ya que i va ir tomando los valores de los ángulos que voy a necesitar (10 en total)
for i in range(0,360,36):
        y=(0.1*np.sin(i))##Calculo la coordenada polar en "y"
        x=(0.1*np.cos(i)) ##Calculo la coordenada polar en "x"
        ##Imprimo las coordenadas de cada punto, los cuales forman un círculo con respecto a su centro
        print(x)
        print(y) 
        plt.scatter(x,y,color='orangered')##Imprimo la gráfica
```

```python
##Este FOR corre 10 veces, y va desde 0 hasta 9
for k in range (10):
    ##Con este FOR asigno a las variables u y l, los valores que se encuentran en las coordenadas que van cambiando con k del arreglo "z" 
    u=z[0,k]
    l=z[1,k]
    plt.scatter(u,l, color='blue') ##Imprimo los centros de los círculos, los cuales están ubicados de forma aleatoria
    if k<5: ##Pongo esta condición, para que sólo se calculen los puntos que forma un círculo alrededor de 5 puntos aleatorios
        for i in range(0,360,36):##Hago el cálculo de las coordenadas polares para obtener sólo 10 de ellas
            y=((0.1)*np.sin(i))
            x=((0.1)*np.cos(i))
            plt.scatter(x+u,y+l,color='orangered')##Imprimo la gráfica que se obtiene
        
     
```
