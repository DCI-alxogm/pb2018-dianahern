# EJERCICIO OSCILADOR

```python
import numpy as np ##Primero hago la declaración de mis librerías
import matplotlib.pyplot as plt##Numpy sirve para los arreglos y las funciones básicas, matplotlib para graficar
%matplotlib inline
```

```python
##Pongo los valores iniciales que son fijos, los cuales pueden ser modificados por el usuario
w=0.5
h=0.0001
tmax=100
vi=0.5
```

# FUNCIÓN INICIALIZAR

```python
def fx(h,tmax,vi):##Esta función tiene como argumentos de entrada los valores de h,x y vi
    t=np.arange(0,tmax,h) ##T es un arreglo que inicia en 0, llega hasta el valor de tmax y va aumentando de h en h
    print (t)
    v=np.zeros(len(t)) ##v es un arreglo con elementos iguales a t pero cuyos valores son 0
    v[0]=vi
    print(v)
    theta=np.zeros(len(t))##theta es un arreglo con elementos iguales a t pero cuyos valores son 0
    print(theta)
    return (t,v,theta) ##Se regresan los valores de los 3 arreglos
```

```python
t,v,theta= fx(h,tmax,vi) ##Llamo a mi función poniendole sus argumentos de entrada
aux= len (t) ##Le asigno una variable a la longitud de tiempo, el cual voy a usar para mi FOR en la siguiente función
```

# FUNCIÓN CÁLCULOS

```python
def fx2(v,h,theta,aux,w):##Esta función tiene como argumentos de entrada los valores de v,h,theta,aux y w
    for i in range (aux-1):##Este for ayuda a rellenar los arreglos de v y theta en cada posición
        v[i+1]=v[i]+h*(-w**2*theta[i])##Hago las operaciones 
        theta[i+1]=theta[i]+h*v[i]
    return (v,theta) ##Regreso los arreglos ya completos y con nuevos valores de v y theta
```

```python
v,theta=fx2(v,h,theta,aux,w) ##Llamo la función2 poniendo los argumentos de entrada, como estos no cambian, sólo pongo las variables que ya había declarado anteriormente
```

# GRÁFICAS

```python
plt.plot(t,theta, color='r')##Grafico tiempo sobre theta
plt.xlabel('Tiempo') ##Pongo los labels, títulos de los gráficos
plt.ylabel('Ángulo')
```

```python
plt.plot(t,v, color='violet')##Grafico tiempo sobre velocidad
plt.xlabel('Tiempo')##Pongo los labels, títulos de los gráficos
plt.ylabel('Velocidad')
```

```python
plt.plot(v,theta,color='blue')##Grafico velocidad sobre ángulo
plt.xlabel('Velocidad')##Pongo los labels, títulos de los gráficos
plt.ylabel('Ángulo')
```
