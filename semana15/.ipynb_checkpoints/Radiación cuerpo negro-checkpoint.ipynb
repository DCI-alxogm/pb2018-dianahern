# Radiación de un cuerpo negro

```python
import numpy as np ##Hago la declaración de la librería general
import matplotlib.pyplot as plt ##Declaro la librería que sirve para graficar
import scipy.constants as cte ##Declaro esta librería que sirve para obtener constantes

%matplotlib inline
```

```python
datos=np.loadtxt('IRCF.txt') ##SE leen los datos de mi archivo "IRCF.txt" gracias a la función np.loadtxt y se ponen en una variable de tipo arreglo
```

# Variables usadas

```python
##Hago la declaración de otrtas variables y obtengo los valores de las constantes conocidas gracias a la librería de scipy.constants
h=cte.Planck 
k=cte.Boltzmann
c=cte.c

frecuencia=datos[:,0] ##Se le otorga una variable a la columna #1 
intensidad=datos[:,1] ##Se le otorga una variable a la columna #2
intens=intensidad ##Utilizo esta variable para poder graficar la intensidad original
barras=datos[:,3] ##Se le otorga una variable a la columna #4

barrasMEGAS=barras*1e-3 ##Hago esto para convertir Megas en KIlo Jankys
y=np.linspace(0,0,(len(intensidad))) ##Esta variable sirve para dar 43 puntos en y y poder graficar barras de error
error=np.linspace(0,0,(len(intensidad))) ##
t=np.linspace(1,5,100)
x2=4000
x=0
treal=0
```

# Gráfica de los datos obtenidos en archivo de texto

```python
plt.subplot(2,1,1) #Esto indica que es la gráfica de arriba
plt.plot(frecuencia, intensidad, color='black', marker='o')
plt.xlabel('Frecuencia (cm^-1)') ##Pongo los labels en x y y
plt.ylabel('Intensidad (MJy/sr)') 
plt.title ('Rad. fondo microondas') ##POngo el titulo de la gráfica


plt.subplot(2,1,2) #Esto indica que es la gráfica de abajo
plt.errorbar(frecuencia,y, barrasMEGAS) #Grafico las barras de error
plt.xlabel('Frecuencia (cm^-1)') ##Pongo los labels en x y y
plt.ylabel('Incertidumbre (MJy/sr)')
```

# Cálculo de las intensidaddes en función de 10 temperaturas

```python
for j in range(len(t)): #Se hace un for para que recorra todos los valores de la temperatura (100)
    x=0
    o=fx(frecuencia, t[j]) #Se llama a la función fx para que calcule la intensidad con cada temperatura
    ##o=intensidad de esa temperatura
    
#La siguiente parte sirve para graficar las intensidades de las temperaturas con el menor margen de error porsible
    for l in range(len(barras)): #Se hace un FOR que tenga la longitud de las barras de error (43)
        error[l]=((intensidad[l]-o[l])/barrasMEGAS[l])**2 #SE declara una variable de tipo arreglo que calcula el error mínimo 
        x=x+error[l] #Se va sumando el error al que ya se tenía antes
     
    if(x<x2):#Condición para disminuir el margen de error entre todas las intensidades
        x2=x
        treal=t[j] 
           
    if(j%10==0): 
        grafica(frecuencia,o,intens) #Se llama la función GRÁFICA para graficar cada 10 temperaturas
 
```

# Función para calcular intensidad 

```python
def fx(frecuencia,t):
    f=frecuencia*c*10**2 #Se convierten las unidades de cm a hertz
    return((h*pow(f,3)))/(pow(c,2))*(1/((np.exp((h*f)/(k*t)))-1))*(10**20) ##Me está regresando la intensidad calculada con las diferentes temperaturas
#Al final se multiplica por 10e20 para cambiar de SI a unidades de Jansky
```

# Función para graficar

```python
def grafica(frecuencia,intensidad,intens):##Tiene 3 elementos de entrada, que es la frecuencia (nunca cambia), o (la nueva intensidad calculada) e intens (que es el valor de intensidad del cual se parte)
    plt.subplot(2,1,1) #Esto indica que es la gráfica de arriba
    plt.plot(frecuencia, intens, color='black', marker='o') #Se grafica la gráfica con datos originales con puntos
    plt.plot(frecuencia, intensidad) #Se grafican las demás gráficas con diferentes temperaturas
    plt.ylabel('Intensidad (MJy/sr)')#POngo unidades de la intensidad en eje y
    plt.title ('Rad. fondo microondas') #POngo el titulo de la gráfica


    plt.subplot(2,1,2)#Esto indica que es la gráfica de abajo
    plt.errorbar(frecuencia,y, barrasMEGAS) #Grafico las barras de error 
    plt.xlabel('Frecuencia (cm^-1)') ##Pongo los labels en x 
    plt.ylabel('Incertidumbre (MJy/sr)') #POngo unidades de la intensidad en eje y     
```
